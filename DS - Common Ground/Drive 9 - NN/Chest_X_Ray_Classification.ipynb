## Constants

```python
INPUT_TRAIN_DIR_PATH = "chest_xray/train/"
INPUT_VAL_DIR_PATH = "chest_xray/val/"
INPUT_TEST_DIR_PATH = "chest_xray/test/"

NORMAL_DIR = "NORMAL/"
PNEUMONIA_DIR = "PNEUMONIA/"


DATASET_TYPE_DICT = {'train':INPUT_TRAIN_DIR_PATH, 'val':INPUT_VAL_DIR_PATH, 'test':INPUT_TEST_DIR_PATH}
TARGET_TYPE_DICT = {'normal':NORMAL_DIR, 'pneumonia':PNEUMONIA_DIR}


IMAGE_HEIGHT = 256
IMAGE_WIDTH = 256
IMAGE_CHANNELS = 1
VGG_MEAN_NORM = sum([0.485*0.3, 0.456*0.59, 0.406*0.11]) # based on https://pytorch.org/hub/pytorch_vision_vgg/
VGG_STD_NORM = sum([0.229*0.3, 0.224*0.59, 0.225*0.11]) # based on https://pytorch.org/hub/pytorch_vision_vgg/
CLASS_NUMBER = 2
TRAIN_SIZE = 0.8

ROTATE_DEGREE = 15 # per side (from -15 to 15)
HEIGHT_CROP = 0.2 # per side
WIDTH_CROP = 0.1 # per side

BATCH_SIZE = 16
LEARNING_RATE = 0.005612 # started with 0.01
MOMENTUM = 0.630581 # Started 0.8
MAX_EPOCHS = 50


# For later use
torch.set_default_tensor_type(torch.cuda.FloatTensor)
device = "cuda" if torch.cuda.is_available() else "cpu"
model = torch.hub.load('pytorch/vision:v0.6.0', 'vgg16', pretrained=True)
```

## Functions

```python
# Redundant...
def split_pneumonia_file(path):
    """
    Create a dir for each class (for later use in load_data)
    """
    os.makedirs(path+VIRUS_DIR, exist_ok=True)
    os.makedirs(path+BACTERIA_DIR, exist_ok=True)
    
    for infile in glob.iglob(path+PNEUMONIA_DIR+"/*"):
        if infile.split('_')[2] == 'virus':
            os.rename(infile, path+VIRUS_DIR+infile.split('/')[-1])
        elif infile.split('_')[2] == 'bacteria':
            os.rename(infile, path+BACTERIA_DIR+infile.split('/')[-1])
        else:
            raise Exception(f'Rough file: {infile}')
    if os.path.isdir(path+PNEUMONIA_DIR):
        os.removedirs(path+PNEUMONIA_DIR)
        

# Heavily based on https://www.programcreek.com/python/example/104833/torchvision.transforms.ToTensor
def load_data(data_folder, batch_size, shuffle=False, dataset_type=None):
    """
    Create a dataloader for train/val/test
    Assumes the filder has 3 subfolders (one per class)
    Resizes and transform to tensor
    """
    transforms = {'train': torchvision.transforms.Compose(
        [torchvision.transforms.RandomRotation(ROTATE_DEGREE),
         torchvision.transforms.Resize([IMAGE_HEIGHT, IMAGE_WIDTH]),
         torchvision.transforms.CenterCrop((IMAGE_HEIGHT * (1-2*HEIGHT_CROP), 
                                            IMAGE_WIDTH * (1-2*WIDTH_CROP))),
         torchvision.transforms.RandomHorizontalFlip(),
         torchvision.transforms.ToTensor(),
         torchvision.transforms.Normalize(mean=VGG_MEAN_NORM, std=VGG_STD_NORM)]),
                  'val': torchvision.transforms.Compose(
        [torchvision.transforms.Grayscale(num_output_channels=IMAGE_CHANNELS),
         torchvision.transforms.Resize([IMAGE_HEIGHT, IMAGE_WIDTH]),
         torchvision.transforms.CenterCrop((IMAGE_HEIGHT * (1-2*HEIGHT_CROP), IMAGE_WIDTH * (1-2*WIDTH_CROP))),
         torchvision.transforms.Resize([IMAGE_HEIGHT, IMAGE_WIDTH]), # For good input size
         torchvision.transforms.ToTensor(),
         torchvision.transforms.Normalize(mean=VGG_MEAN_NORM, std=VGG_STD_NORM)])}
    # ImageFolder assumes the data in the folder is divided to subfolders using classes 
    # e.g. folder for normal, virus, bacteria
    data = torchvision.datasets.ImageFolder(root = data_folder, transform=transforms[dataset_type])
    
    data_loader = DataLoader(data, batch_size=batch_size, shuffle=shuffle)
    return data_loader 


def get_data_loaders(batch_size, train_size=TRAIN_SIZE):
    """
    Create dataloaders for train/val
    Assumes the filder has 3 subfolders (one per class)
    Resizes and transform to tensor
    """
    transforms = torchvision.transforms.Compose(
        [torchvision.transforms.Grayscale(num_output_channels=IMAGE_CHANNELS),
         torchvision.transforms.RandomRotation(ROTATE_DEGREE),
         torchvision.transforms.Resize([IMAGE_HEIGHT, IMAGE_WIDTH]),
         torchvision.transforms.CenterCrop((IMAGE_HEIGHT * (1-2*HEIGHT_CROP), IMAGE_WIDTH * (1-2*WIDTH_CROP))),
         torchvision.transforms.Resize([IMAGE_HEIGHT, IMAGE_WIDTH]), # For good input size
         torchvision.transforms.RandomHorizontalFlip(),
         torchvision.transforms.ToTensor(),
         torchvision.transforms.Normalize(mean=VGG_MEAN_NORM, std=VGG_STD_NORM)])
    
    data_from_train = torchvision.datasets.ImageFolder(root = INPUT_TRAIN_DIR_PATH, transform=transforms)
    data_from_val = torchvision.datasets.ImageFolder(root = INPUT_VAL_DIR_PATH, transform=transforms)

    full_dataset = ConcatDataset([data_from_train,data_from_val])

    train_data, val_data = random_split(full_dataset, [int(len(full_dataset)*TRAIN_SIZE), 
                                                       len(full_dataset)-int(len(full_dataset)*TRAIN_SIZE)])


    train_data_loader = DataLoader(train_data, batch_size=batch_size)
    val_data_loader = DataLoader(val_data, batch_size=batch_size)
    
    return train_data_loader,  val_data_loader
    
```

## Data & Model Preperation

```python
train_data_loader, validation_data_loader = get_data_loaders(batch_size=BATCH_SIZE, train_size=TRAIN_SIZE)
# train_data_loader = load_data(INPUT_TRAIN_DIR_PATH, BATCH_SIZE, shuffle=True, dataset_type='train')
print(f"Have {len(train_data_loader)} batches in the training set")
# validation_data_loader = load_data(INPUT_VAL_DIR_PATH, BATCH_SIZE, shuffle=False, dataset_type='val')
print(f"Have {len(validation_data_loader)} batches in the validation set")
```

```python
# Probably redundant but  make sure all weights will be trained
for param in model.features.parameters():
    param.requires_grad = True
for param in model.classifier.parameters():
    param.requires_grad = True

# edit first layer - change number of input channels to 1
model.features[0] = nn.Conv2d(1,64,kernel_size=(3,3), stride=(1,1), padding=(1,1))

features = list(model.classifier.children())[:-1] # Remove last layer
features.extend([nn.Linear(model.classifier[-1].in_features, CLASS_NUMBER)]) # Add layer
features.extend([nn.Softmax(dim=1)]) # Add layer

model.classifier = nn.Sequential(*features) # Replace the model classifier
print(model)
```

## Fit Model

### "Basic Fitting" - Fun with Tensorboard

```python
os.makedirs('./tensorboard', exist_ok=True)
writer = SummaryWriter(log_dir="./tensorboard")

criterion = nn.CrossEntropyLoss()
optimizer = torch.optim.SGD(model.parameters(), lr=LEARNING_RATE, momentum=MOMENTUM)

for epoch in tqdm(range(MAX_EPOCHS)):
    preds = []
    preds_proba = []
    true_labels = []
    for i, (X_train, y_train) in enumerate(train_data_loader):
        # Save real labels
        curr_true_labels = [el.item() for el in y_train]
        true_labels = true_labels+curr_true_labels
        # Get and save predicted labels
        curr_preds = model(X_train.type(torch.cuda.FloatTensor))
        # high thresh for more representive accuracy
        preds = preds + [1 if el[1]>0.995 else 0 for el in curr_preds.cpu().detach().numpy()]
        preds_proba.append(curr_preds)
        
        curr_train_loss = criterion(curr_preds, y_train) # NOTE: preds should be first
        # high thresh for more representive accuracy
        curr_train_accuracy = accuracy_score([1 if el[1]>0.995 else 0 for el in curr_preds.cpu().detach().numpy()], 
                                             curr_true_labels)
        print(f"debug: curr_train_loss={curr_train_loss} curr_train_accuracy={curr_train_accuracy}")
        
        # Write metrics to tensofrflow
        writer.add_scalar("loss/train_loss", curr_train_loss, 
                           global_step=epoch*len(train_data_loader)+i)
        writer.add_scalar('accuracy/train_accuracy', curr_train_accuracy,
                           global_step=epoch*len(train_data_loader)+i)
        
        # Update model's params
        curr_train_loss.backward()
        optimizer.step()
        optimizer.zero_grad()
        
    fig = plt.figure()
    sns.heatmap(confusion_matrix(y_pred=preds, y_true=true_labels, normalize='true'), cmap='Greens', annot=True)
    writer.add_figure("Train Confusion Matrix", fig, global_step=epoch)
    
    writer.add_pr_curve(tag="Train PR Curve", 
                            labels=torch.tensor(true_labels), 
                        # TODO: simplify next line (done during debug...)
                            predictions=torch.tensor([el[1] for el in torch.cat(preds_proba).cpu().detach().numpy()]), 
                            global_step=epoch)
    
    # Get val scores
    with torch.no_grad():
        val_real = torch.tensor([])
        val_preds = torch.tensor([])
        for i, (X_val, y_val) in enumerate(validation_data_loader):
            val_real = torch.cat([val_real, y_val.type(torch.cuda.FloatTensor)])
            curr_val_preds = model(X_val.type(torch.cuda.FloatTensor))
            val_preds = torch.cat([val_preds, curr_val_preds])
            
            # Add images to tensorboard
            is_correct = [pred == true_label for pred,true_label in zip([1 if el[1]>0.995 else 0 for el in val_preds.cpu().detach().numpy()], 
                                                                    y_val.type(torch.cuda.FloatTensor))]
            try:
                idx_of_misclassified = is_correct.index(False)
                writer.add_image("Missclassified Examples", X_val[idx_of_misclassified], global_step=epoch)
            except:
                print("No bad example in this iteration")
            
        writer.add_pr_curve(tag="Val PR Curve", 
                                labels=val_real, 
                                predictions=torch.tensor([el[1] for el in val_preds.cpu().detach().numpy()]), 
                                global_step=epoch)
        fig = plt.figure()
        # high thresh for more representive conf matrix
        sns.heatmap(confusion_matrix(y_pred=[1 if el[1]>0.995 else 0 for el in val_preds.cpu().detach().numpy()], 
                                     y_true=val_real.cpu(), normalize='true'), cmap='Greens', annot=True)
        writer.add_figure("Val Confusion Matrix", fig, global_step=epoch)
        
```

### Hyper-Param Search (Optuna)

```python
def train_vgg(trial):
    
    cfg={
        'lr'       : trial.suggest_loguniform('lr', 0.001, 0.01),          
        'momentum' : trial.suggest_uniform('momentum', 0.2, 0.99),
        'weight_decay' : trial.suggest_loguniform('weight_decay', 0.01, 0.1), # was accidentaly 'lr' so... woops
    }
    
    model = torch.hub.load('pytorch/vision:v0.6.0', 'vgg16', pretrained=True)
    # Probably redundant but  make sure all weights will be trained
    for param in model.features.parameters():
        param.requires_grad = True
    for param in model.classifier.parameters():
        param.requires_grad = True
    # edit first layer - change number of input channels to 1
    model.features[0] = nn.Conv2d(1,64,kernel_size=(3,3), stride=(1,1), padding=(1,1))
    features = list(model.classifier.children())[:-1] # Remove last layer
    features.extend([nn.Linear(model.classifier[-1].in_features, CLASS_NUMBER)]) # Add layer
    features.extend([nn.Softmax(dim=1)]) # Add layer
    model.classifier = nn.Sequential(*features) # Replace the model classifier
    
    criterion = nn.CrossEntropyLoss()
    optimizer = torch.optim.SGD(model.parameters(), lr=cfg['lr'], 
                                momentum=cfg['momentum'], 
                                weight_decay=cfg['weight_decay'])
    
    for epoch in tqdm(range(10)): # reduced to 10 for reducing time
        for i, (X_train, y_train) in enumerate(train_data_loader):
            # Save real labels
            curr_true_labels = [el.item() for el in y_train]
            # Get and save predicted labels
            curr_preds = model(X_train.type(torch.cuda.FloatTensor))
            curr_train_loss = criterion(curr_preds, y_train) # NOTE: preds should be first

            # Update model's params
            curr_train_loss.backward()
            optimizer.step()
            optimizer.zero_grad()

    # Get val scores
    with torch.no_grad():
        val_real = torch.tensor([])
        val_preds = torch.tensor([])
        for i, (X_val, y_val) in enumerate(validation_data_loader):
            val_real = torch.cat([val_real, y_val.type(torch.cuda.FloatTensor)])
            curr_val_preds = model(X_val.type(torch.cuda.FloatTensor))
            val_preds = torch.cat([val_preds, curr_val_preds])
    
    val_loss = criterion(val_preds, val_real.type(torch.long))
    print(f"final train loss = {val_loss}")
    return val_loss
            
            
            
            
sampler = optuna.samplers.TPESampler()
study = optuna.create_study(sampler=sampler)
study.optimize(train_vgg, n_trials=50)
joblib.dump(study, 'optuna.pkl')
```

## Test Evaluation

```python
test_data_loader = load_data(INPUT_TEST_DIR_PATH, BATCH_SIZE, shuffle=False, dataset_type='val')
print(f"Have {len(test_data_loader)} batches in the test set")
```

```python
pred = []
real = []
thresh = 0.99999999 # High - we prefer less FN (healthy classified as sick) 
with torch.no_grad():
    for batch in iter(test_data_loader):
        batch[0] = batch[0].type(torch.cuda.FloatTensor)
        batch_pred = model(batch[0])
        real = real + [el.item() for el in batch[1]]
        pred = pred + [1 if el[1]>thresh else 0 for el in batch_pred.cpu().numpy()]
```

```python
print(f"Accuracy = {accuracy_score(pred, real)}")
fig = sns.heatmap(confusion_matrix(y_pred=pred, y_true=real, normalize='true'), cmap='Greens', annot=True)
fig.set(xlabel='Predicted', ylabel='Real', title="Test Confusion Matrix")
```
