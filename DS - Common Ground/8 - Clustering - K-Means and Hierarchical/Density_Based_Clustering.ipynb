# Density Based Clustering
Cluster a given dataset using DBSCAN and mark outliers and stuff

## Imports

```python
import pandas as pd
import numpy as np
from sklearn.cluster import DBSCAN
from sklearn.datasets import make_blobs
import plotly.express as px
```

## Consts

```python
INPUT_PATH = "" # Missing: data.csv file
COLUMNS = ['x', 'y', 'class']
TARGET = 'class'

RANDOM_SEED = 6
N_CLUSTERS = 3 # For dataset creation
N_SAMPLES = 300 # For dataset creation
N_FEATURES = 2 # For dataset creation
```

## Read the Data
NOTE: no data file available, will create a similar df for convinience

```python
X, y = make_blobs(n_samples=N_SAMPLES, centers=N_CLUSTERS, n_features=N_FEATURES,random_state=RANDOM_SEED)
df = pd.DataFrame(np.column_stack((X,y)), columns=COLUMNS)
```

```python
fig = px.scatter(df, x='x', y='y', color='class')
fig.update_layout(title = 'Data Scatter Plot Using Real Clusters')
fig.show()
```

## Clustering

```python
dbscan = DBSCAN() # Default parameters for now...
res = dbscan.fit_predict(df.drop(TARGET, axis=1))
```

```python
fig = px.scatter(df, x='x', y='y', color=res)
fig.update_layout(title = 'Data Scatter Plot Using Predicted Clusters')
fig.show()
```
