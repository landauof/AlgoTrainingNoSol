# Random Forest Example
Demonstrate in code that when the number of variables is large, but the fraction of relevant variables is small, random  forests are likely to perform poorly with small m.

```python
import numpy as pd
import numpy as np
from tqdm import tqdm

import plotly.graph_objects as go

from sklearn.datasets import load_iris
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score
```

```python
X,y = load_iris(return_X_y=True)
```

```python
accuracies = []
noise_size = []
for random_col_number in tqdm(range(0,1000,5)):
    accuracies.append(np.average(cross_val_score(RandomForestClassifier(),
                                                 np.append(X, np.random.rand(len(X), random_col_number), axis=1),
                                                 y)))
    noise_size.append(random_col_number)
```

```python
fig = go.Figure(data=go.Scatter(x=noise_size, y=accuracies))
fig.update_layout(
    title="Accuracy per Noise (3 informative columns)",
    xaxis_title="Number of noise columns",
    yaxis_title="Accuracy",
)
fig.show()
```
